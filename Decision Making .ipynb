```python
import pandas as pd
df = pd.read_csv("C:/Users/shiva/Downloads/Company Stratergies.csv")
df
```

```python
# Check for missing values
print(df.isnull().sum())

# Fill missing values if any (example: filling with mean of the column)
# df.fillna(df.mean(), inplace=True)

# Verify if any missing values remain
print(df.isnull().sum())
```

```python
# Calculate annual growth rates
df['Revenue Growth Rate (%)'] = df['Total Revenue ($M)'].pct_change() * 100
df['Product A Sales Growth (%)'] = df['Product A Sales (units)'].pct_change() * 100
df['Product B Sales Growth (%)'] = df['Product B Sales (units)'].pct_change() * 100
df
```

```python
# Fill NaN values with 0% (assuming no growth for the first year)
df['Revenue Growth Rate (%)'].fillna(0, inplace=True)
df['Product A Sales Growth (%)'].fillna(0, inplace=True)
df['Product B Sales Growth (%)'].fillna(0, inplace=True)

# Display the data with filled NaN values
print(df[['Year', 'Revenue Growth Rate (%)', 'Product A Sales Growth (%)', 'Product B Sales Growth (%)']])
```

```python
import matplotlib.pyplot as plt

# Plot revenue trend
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Total Revenue ($M)'], marker='o', label='Total Revenue')
plt.xlabel('Year')
plt.ylabel('Total Revenue ($M)')
plt.title('Total Revenue Trend')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Summary statistics
print(df.describe())
```

```python
# Calculate correlation matrix
correlation_matrix = df.corr()
print(correlation_matrix)
```

```python
import statsmodels.api as sm

# Example: Regression of Revenue on R&D Expenditure
X = df['R&D Expenditure ($M)']
y = df['Total Revenue ($M)']
X = sm.add_constant(X)  # Add a constant term for intercept

model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Example: Forecasting Total Revenue
model = ExponentialSmoothing(df['Total Revenue ($M)'], trend='add', seasonal=None, seasonal_periods=1)
fit = model.fit()
forecast = fit.forecast(steps=3)  # Forecasting next 3 years
print(forecast)
```

```python
# Revenue vs. Market Share
import seaborn as sns
plt.figure(figsize=(12, 8))
sns.lineplot(data=df, x='Year', y='Total Revenue ($M)', marker='o', label='Total Revenue')
sns.lineplot(data=df, x='Year', y='Market Share (%)', marker='o', label='Market Share (%)', color='orange')
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Revenue and Market Share Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
fig, axs = plt.subplots(3, 2, figsize=(14, 12))

# Plot each metric
axs[0, 0].plot(df['Year'], df['Total Revenue ($M)'], marker='o', color='b')
axs[0, 0].set_title('Total Revenue')

axs[0, 1].bar(df['Year'], df['Product A Sales (units)'], color='g')
axs[0, 1].set_title('Product A Sales')

axs[1, 0].bar(df['Year'], df['Product B Sales (units)'], color='r')
axs[1, 0].set_title('Product B Sales')

axs[1, 1].plot(df['Year'], df['Market Share (%)'], marker='o', color='purple')
axs[1, 1].set_title('Market Share')

axs[2, 0].plot(df['Year'], df['Customer Satisfaction (%)'], marker='o', color='cyan')
axs[2, 0].set_title('Customer Satisfaction')

axs[2, 1].plot(df['Year'], df['ROI (%)'], marker='o', color='magenta')
axs[2, 1].set_title('ROI')

for ax in axs.flat:
    ax.set_xlabel('Year')
    ax.set_ylabel('Value')
    ax.grid(True)

plt.tight_layout()
plt.show()
```
